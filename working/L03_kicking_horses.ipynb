```python
import numpy as np
from matplotlib import pyplot as plt
import math

plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.serif'] = ['Liberation Serif'] + plt.rcParams['font.serif']
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=[8,6]
```

```python
tot_g =200 #total group number

#data
n_g=np.array([109,65,22,3,1])
n_d=np.array([0,1,2,3,4])

plt.plot(n_d, n_g, label='data', lw=3)
plt.xlabel('deaths')
plt.ylabel('groups')
plt.legend()
plt.show()
```

```python
#compute the mean
tot_d=0
for i in range(len(n_g)):
    tot_d+=n_g[i]*n_d[i]

mu=tot_d/tot_g
print("mean: ", mu)

#compute the poissonian distribution for that mean
def fun(x):
    return (mu**x*np.exp(-mu))/math.factorial(x)

y=np.empty(len(n_g))

for i in range(len(n_g)):
    y[i]=fun(n_d[i])

print("fraction of groups from data:", n_g/tot_g)
print("poissonian results: ", y)

plt.plot(n_d,n_g/tot_g, label='data', lw=3)
plt.plot(n_d,y, label='poissonian', lw=3, ls='--')
plt.xlabel('deaths')
plt.ylabel('groups')
plt.legend()
plt.show()
```
