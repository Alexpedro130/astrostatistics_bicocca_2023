### 2. Kicking horses

A famous early  application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by [Bortkiewicz](https://www.wikiwand.com/en/Ladislaus_Bortkiewicz) in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:

| Number of deaths | Number of groups |
| --- | --- |
| 0 | 109 |
| 1 | 65 |
| 2 | 22 |
| 3 | 3 |
| 4 | 1 |

- Plot the resulting probability distribution. Careful with the normalization
- How does it look like? Again try to fit it by hand.

```python
import numpy as np
from matplotlib import pyplot as plt
import math

plt.rcParams['font.family'] = 'serif'
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.serif'] = ['Liberation Serif'] + plt.rcParams['font.serif']
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=[6,4]
```

```python
tot_g =200 #total group number

#data
n_g=np.array([109,65,22,3,1])
n_d=np.array([0,1,2,3,4])

plt.plot(n_d, n_g, label='data', lw=2)
plt.xlabel('deaths')
plt.ylabel('groups')
plt.legend()
plt.show()
```

Try to manually fit a poissonian distribution

```python
#compute the mean
tot_d=0
for i in range(len(n_g)):
    tot_d+=n_g[i]*n_d[i]

mu=tot_d/tot_g
print("mean: ", mu)

#compute the poissonian distribution for that mean
def fun(x):
    return (mu**x*np.exp(-mu))/math.factorial(x)

y=np.empty(len(n_g))

for i in range(len(n_g)):
    y[i]=fun(n_d[i])

print("fraction of groups from data:", n_g/tot_g)
print("poissonian results: ", y)

plt.plot(n_d,n_g/tot_g, label='data', lw=2, alpha=0.8)
plt.plot(n_d,y, label='poissonian', lw=2, ls='--')
plt.xlabel('deaths')
plt.ylabel('groups')
plt.legend()
plt.show()
```

The fit follows the distribution quite well
